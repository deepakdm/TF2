## IMPORT LIBRARIES

```python
import tensorflow as tf
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## IMPORT DATASETS

```python
temp_df = pd.read_csv('Celsius-to-Fahrenheit.csv')
```

## VISUALIZE DATASET

```python
sns.scatterplot(temp_df['Celsius'], temp_df['Fahrenheit'])
```

## CREATE TESTING AND TRAINING DATASET

```python
X_train = temp_df['Celsius']
y_train = temp_df['Fahrenheit']
```

## BUILD AND TRAIN THE MODEL

```python
X_train.shape
```

```python
y_train.shape
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units=1, input_shape=[1]))
```

```python
model.summary()
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(0.5), loss='mean_squared_error')
```

```python
epochs_hist = model.fit(X_train, y_train, epochs = 500)
```

## EVALUATING THE MODEL

```python
epochs_hist.history.keys()
```

```python
plt.plot(epochs_hist.history['loss'])
plt.title('Model Loss Progress During Training')
plt.xlabel('Epoch')
plt.ylabel('Training Loss')
plt.legend(['Training Loss'])
```

```python
model.get_weights()
```

```python
# Use the trained model to perform predictions

Temp_C = 0
Temp_F = model.predict([Temp_C])
print('Temperature in degF Using Trained ANN =', Temp_F)
```

```python
# Let's confirm this Using the equation: 
Temp_F = 9/5 * Temp_C + 32
print('Temperature in degF Using Equation =', Temp_F)
```
